3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Создадим функцию преобразования
import re
def parse_experience(value):
    if pd.isna(value) or 'не указано' in value.lower():
        return np.nan
    
    # Ищем количество лет и месяцев
    years = 0
    months = 0
    
    # Ищем "n лет" или "n год/года/лет"
    match_years = re.search(r'(\d+)\s+год', value)
    if match_years:
        years = int(match_years.group(1))
    
    # Ищем "m месяц/месяца/месяцев"
    match_months = re.search(r'(\d+)\s+месяц', value)
    if match_months:
        months = int(match_months.group(1))
        
    return years * 12 + months

# Применим функцию к столбцу
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(parse_experience)

# Удалим оригинальный столбец
df.drop(columns=['Опыт работы'], inplace=True)

# Проверим результат
print(df['Опыт работы (месяц)'].describe())
print(df['Опыт работы (месяц)'].isna().sum())  # Кол-во пропусков
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Список городов-миллионников
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск',
                  'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 
                  'Воронеж','Волгоград']

# Функция для извлечения города
def extract_city(value):
    city = value.split(',')[0].strip()
    if city == 'Москва':
        return 'Москва'
    elif city == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    elif city in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

# Функция для извлечения готовности к переезду
def extract_relocation(value):
    value = value.lower()
    return any(x in value for x in ['готов к переезду', 'готова к переезду', 'хочу переехать'])

# Функция для извлечения готовности к командировкам
def extract_business_trip(value):
    value = value.lower()
    if 'не готов' in value and 'командиров' in value:
        return False
    elif 'командиров' in value:
        return True
    else:
        # Если про командировки ничего не сказано — считаем False
        return False

# Применяем преобразования
df['Город'] = df['Город, переезд, командировки'].apply(extract_city)
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(extract_relocation)
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(extract_business_trip)

# Удаляем исходный столбец
df.drop(columns=['Город, переезд, командировки'], inplace=True)

# Проверим результат
print(df[['Город', 'Готовность к переезду', 'Готовность к командировкам']].head())
print(df['Город'].value_counts())
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Список категорий для признака "Занятость"
employment_categories = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']

# Список категорий для признака "График"
schedule_categories = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']

# Функция для выполнения One Hot Encoding для категорий
def apply_ohe(value, categories):
    return [cat in value for cat in categories]

# Применяем One Hot Encoding для признака "Занятость"
employment_ohe = df['Занятость'].apply(lambda x: apply_ohe(x, employment_categories))

# Разбиваем полученные списки на отдельные столбцы для каждой категории
for i, cat in enumerate(employment_categories):
    df[cat] = [item[i] for item in employment_ohe]

# Применяем One Hot Encoding для признака "График"
schedule_ohe = df['График'].apply(lambda x: apply_ohe(x, schedule_categories))

# Разбиваем полученные списки на отдельные столбцы для каждой категории
for i, cat in enumerate(schedule_categories):
    df[cat] = [item[i] for item in schedule_ohe]

# Удаляем оригинальные столбцы
df.drop(columns=['Занятость', 'График'], inplace=True)

# Проверим результат
print(df[['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']].head())
print(df[['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']].head())
```

3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Самостоятельно скачали файл с mfd.ru с курсами валют за указанный период
# Прочитаем Excel-файл с указанием разделителя
currency_df = pd.read_excel('data/path_to_currency_data.xltx', dtype=str)

# Разделим строки по символу ";"
currency_df = currency_df.iloc[:, 0].str.split(";", expand=True)

# Назначим правильные названия столбцам
currency_df.columns = ['currency', 'period', 'date', 'time', 'close', 'vol']

# Преобразуем типы данных
currency_df['date'] = pd.to_datetime(currency_df['date'], format='%Y%m%d')
currency_df['close'] = pd.to_numeric(currency_df['close'], errors='coerce')
currency_df['proportion'] = 1  # Для всех записей пропорция 1

# Оставим только нужные столбцы
currency_df = currency_df[['currency', 'date', 'close', 'proportion']]

# Создадим словарь с пропорциями для каждой валюты
proportion_map = {
    'USD': 1,
    'EUR': 1,
    'UAH': 10,
    'KZT': 100,
    'BYN': 1,
    'RUB': 1,
    'UZS': 10000,
    'KGS': 10,
    'AZN': 1
}

# Обновим столбец proportion на основе currency
currency_df['proportion'] = currency_df['currency'].map(proportion_map)

# Проверим полученные категории
print(currency_df['proportion'].unique())
print(currency_df.head())
```

```python
# 1. Приводим "Обновление резюме" и даты валют к нужному формату
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'], format='%d.%m.%Y %H:%M')
df['date'] = df['Обновление резюме'].dt.date
currency_df['date'] = pd.to_datetime(currency_df['date']).dt.date

# 2. Извлекаем сумму и валюту из строки "ЗП"
df['salary_amount'] = df['ЗП'].str.extract(r'([\d\s]+)')[0].str.replace(' ', '').astype(float)
df['salary_currency'] = df['ЗП'].str.extract(r'([A-Za-zА-Яа-я]+)\.?$')[0].str.upper()

# Словарь перевода валют в ISO
currency_iso_map = {
    'USD': 'USD', 'EUR': 'EUR', 'UAH': 'UAH', 'ГРН': 'UAH', 
    'KZT': 'KZT', 'BYN': 'BYN', 'БЕЛРУБ': 'BYN',
    'RUB': 'RUB', 'РУБ': 'RUB',
    'СУМ': 'UZS', 'UZS': 'UZS', 
    'KGS': 'KGS', 'AZN': 'AZN'
}
df['currency'] = df['salary_currency'].map(currency_iso_map)

# 3. Присоединяем курсы валют (left join)
df = df.merge(currency_df, how='left', on=['currency', 'date'])

# Заполняем RUB: курс и пропорция = 1
df['close'] = df['close'].fillna(1)
df['proportion'] = df['proportion'].fillna(1)

# 4. Рассчитываем зарплату в рублях
df['ЗП (руб)'] = df['salary_amount'] * df['close'] / df['proportion']

# Удаляем лишние столбцы
df.drop(columns=['ЗП', 'salary_amount', 'salary_currency', 'currency', 'close', 'proportion', 'date'], inplace=True)

# --- Проверка результатов ---

# Первые строки
print(df[['Обновление резюме', 'ЗП (руб)']].head(10))

# Проверка на пропуски
missing = df['ЗП (руб)'].isna().sum()
print(f"❗ Пропущенные значения в 'ЗП (руб)': {missing}")

# Диапазон зарплат
print(f" Минимальная ЗП: {df['ЗП (руб)'].min():,.2f} руб.")
print(f" Максимальная ЗП: {df['ЗП (руб)'].max():,.2f} руб.")
```

```python
# Количество пропусков по каждому столбцу
missing_values = df.isnull().sum()

# Выводим только те столбцы, где есть пропуски
missing_values = missing_values[missing_values > 0]

# Сортируем по убыванию количества пропусков
missing_values = missing_values.sort_values(ascending=False)

# Показываем результат
print("Количество пропусков в столбцах:")
print(missing_values)
```

3. Итак, у нас есть пропуски в 3ех столбцах: **"Опыт работы (месяц)"**, **"Последнее/нынешнее место работы"**, **"Последняя/нынешняя должность"**. Поступим следующим образом: удалите строки, где есть пропуск в столбцах с местом работы и должностью. Пропуски в столбце с опытом работы заполните **медианным** значением.

```python
# Удаляем строки с пропусками в местах работы и должностях
df = df.dropna(subset=["Последнее/нынешнее место работы", "Последняя/нынешняя должность"])

# Заполняем пропуски в "Опыт работы (месяц)" медианным значением
median_experience = df["Опыт работы (месяц)"].median()
df["Опыт работы (месяц)"] = df["Опыт работы (месяц)"].fillna(median_experience)

# Проверим, остались ли пропуски
print(df.isnull().sum())
```

```python
# Оставляем только те строки, где ЗП в разумных пределах: от 1000 до 1 000 000 рублей
df = df[(df["ЗП (руб)"] >= 1000) & (df["ЗП (руб)"] <= 1_000_000)]

# Проверим диапазон значений после фильтрации
print(df["ЗП (руб)"].describe())
```
